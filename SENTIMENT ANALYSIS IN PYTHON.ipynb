```python
TM['Pred']=s
```

```python
TM.head()
```

```python
c=0
e=0
f=0
d=0
for i in range(len(TM)):
    if TM['Score'][i]==1:
        e=e+1
        if TM['Score'][i]==TM['Pred'][i]:
            c=c+1
    elif TM['Score'][i]==-1:
        f=f+1
        if TM['Score'][i]==TM['Pred'][i]:
            d=d+1
print(c,d,e,f)
```
